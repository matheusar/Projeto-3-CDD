```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
from math import *
import numpy as np
from scipy.stats import norm, probplot
from scipy import stats
import seaborn as sns
import statsmodels.api as sm
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
```

## Introdução

O objetivo deste projeto é analisar o dataset do jogo FIFA19, onde se encontra todos os atributos, preços e características dos jogadores, e a partir desses e de técnicas de machine learining, como regressão linear e Random Forets Classifier, prever o preço do jogador no mercado virtual, para que gamers possam saber o preço dos jogadores que desejam comprar futuramente.

O dataset foi extraido do link: https://www.kaggle.com/karangadiya/fifa19

## Leitura dos dados

```python
dt = pd.read_csv("data.csv")

dt.head()
```

## Limpando o arquivo (removendo as variáveis que não serão utilizadas nas análises)

```python
dt1= dt.drop(columns=['Unnamed: 0','ID', 'Photo', 'Flag', 'Club Logo', 'Body Type', 'LS', 'ST', 'RS', 'LW', 'RW', 'LF', 'CF', 'RF', 'LAM', 'RAM', 'CAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']).fillna(0)
dt1.head()
```

## Ajustando as colunas "Value", "Wage" e "Release Clause" para virarem variáveis numéricas

```python
valor=[]
for e in dt1['Value']:
    if 'M' in e: 
        M=float(e.replace('M','')[1:])*1000000
        valor.append(M)
    elif 'K' in e:
        K=float(e.replace('K',"")[1:])*1000
        valor.append(K)
    else:
        E=float(e.replace('€',''))
        valor.append(E)

wage = []
for f in dt1['Wage']:
    if 'M' in f: 
        M=float(f.replace('M','')[1:])*1000000
        wage.append(M)
    elif 'K' in f:
        K=float(f.replace('K',"")[1:])*1000
        wage.append(K)
    else:
        E=float(f.replace('€',''))
        wage.append(E)
        
rc = []
for g in dt1['Wage']:
    if 'M' in g: 
        M=float(g.replace('M','')[1:])*1000000
        rc.append(M)
    elif 'K' in g:
        K=float(g.replace('K',"")[1:])*1000
        rc.append(K)
    else:
        E=float(g.replace('€',''))
        rc.append(E)

dt1['value'] = valor
dt1['wage'] = wage
dt1['release clause'] = rc
dt2 = dt1.drop(columns=['Value', 'Wage', 'Release Clause'])
dt2.head()
```

A partir dos gráficos acima, conclui-se que a reputação internacional dos jogadores é um fator importante para o preço deste visto que conforme aumenta o nível da reputação, os jogadores ficam tanto com Overall maior, quanto com preço maior.

## Regressão

```python
# Separando as variáveis que serão usadas nas regressões.

atributos = ['Age', 'Overall', 'Potential', 'Special','International Reputation', 'Weak Foot', 
             'Skill Moves', 'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
            'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl',
           'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
           'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
           'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
           'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving',
           'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'value', 'wage', 'release clause']
atr = dt2.loc[:, atributos]
atr.head()
```

```python
#Função da regressão linear

def regress(X,Y):
    X_cp = sm.add_constant(X)
    model = sm.OLS(Y,X_cp)
    results = model.fit()
    return results
```

```python
# Separação entre as variáveis explicativas e a variável dependente (variável que queremos prever, no caso, 'value', o valor
# dos jogadores no mercado virtual).

Y = atr['value']
X = atr[['Age', 'Overall', 'Potential', 'Special','International Reputation', 'Weak Foot', 
             'Skill Moves', 'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
            'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl',
           'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
           'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
           'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
           'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving',
           'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'wage', 'release clause']]
```

```python
# Primeira regressão

results1 = regress(X, Y)
results1.summary()
```

```python
# Ajustando a regressão anterior para que as todas as variáveis explicativas tenham valor p menor que 10%, ou seja, deixamos
# apenas as variáveis explicativas relevantes para compreendermos o comportamento da variável resposta.

X1 = atr[['Age', 'Overall', 'Potential', 'Special','International Reputation', 'Weak Foot', 
             'Skill Moves', 'Crossing', 'Finishing', 'HeadingAccuracy',
            'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl',
           'Agility', 'Reactions', 'ShotPower', 'Jumping', 'Stamina', 'LongShots',
           'Aggression', 'Interceptions', 'Vision', 'Penalties', 'SlidingTackle',
           'GKHandling','GKReflexes', 'wage', 'release clause']]
results2 = regress(X1, Y)
results2.summary()
```

## Plot 3D

```python
# A partir da realização de vários gráficos 3D com diferentes variáveis independentes, escolhemos as variáveis que melhor 
# explicam a variável dependente a partir do plot 3D.

C = results2.params['const']
b = results2.params["wage"]
w = results2.params['release clause']

def f(x1, x2):
    return C+b*x1 + w*x2

npoints = 100
x1 = np.linspace(X1["wage"].min(), X1["wage"].max(), npoints)
x2 = np.linspace(X1["release clause"].min(), X1["release clause"].max(), npoints)
```

```python
#Plot 3D

x1_xx, x2_yy = np.meshgrid(x1, x2)
z = f(x1_xx, x2_yy)
plt3d = plt.figure().gca(projection="3d")
plt3d.plot_surface(x1_xx, x2_yy, z)
plt3d.scatter(X1["wage"], X1["release clause"], Y , color="r");
```

## Predições

### Separando o Dataset em treinamento e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X1, Y, test_size=0.33, random_state=RANDOM_SEED)
```

```python
#Vendo se o possuem a mesma dimensão

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
#Regressão dos dados de treinamento

results3 = regress(X_train, y_train)
results3.summary()
```

Utilizaremos os métodos abaixo para realizar a análise das predições realizadas por serem métricas de regressão, e não de classificação:

<img src=explained_variance.png width=40%>

### Predição usando Regressão Linear

```python
X_test_cp = sm.add_constant(X_test)
y_pred = results3.predict(X_test_cp)
```

```python
from sklearn import metrics
print('Variância explicada: {}'.format(metrics.explained_variance_score(y_test, y_pred)))
print('R**2: {}'.format(metrics.r2_score(y_test, y_pred)))
print('Erro absoluto médio: {}'.format(metrics.mean_absolute_error(y_test, y_pred)))
```

### Predição usando Random Forest Classifier

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train.astype('int'), y_train.astype('int'))
y_predRFC = model.predict(X_test)
print('Variância explicada: {}'.format(metrics.explained_variance_score(y_test, y_predRFC)))
print('R**2: {}'.format(metrics.r2_score(y_test, y_predRFC)))
print('Erro absoluto médio: {}'.format(metrics.mean_absolute_error(y_test, y_predRFC)))
```

## Conclusão

A partir das métricas utilizadas para validar as predições realizadas, percebe-se que ambas chegaram a valores de mercado próximos aos esperados, já que a variância explicada e o $R^2$ estão bem próximos de 1, ou seja, possuem baixa dispersão e que as variáveis regressoras explicam bem a variável dependente, além do erro absoluto médio ser baixo (levando em consideração que há jogadores que valem 100 milhões de euros), logo os valores encontrados são próximos dos reais. Porém, a Regressão Linear foi menos eficaz que a Random Forest Classifier, de modo que esta chegou à variância explicada e $R^2$ cerca de 6% maior e erro absoluto médio cerca de 881580 euros mais baixo que aquela.
